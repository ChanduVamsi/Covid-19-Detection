```python
#Total number of useful postanterior Covid-19 affected patients from the dataset.
cnt = 0

for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename = row["filename"]
        image_path = os.path.join(IMAGES_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        print("Moving image ",cnt)
        cnt += 1
print(cnt)
```

```python
#So, we have all our Covid +ve XRays in Covid folder of Dataset.
#So, we got 195 XRay images of Covid-19 patients.
#Having a 50%-50% of Covid +ve-Normal gives best results while training.
```

```python
#XRays of Normal Persons.
NORMAL_FILE_PATH = "C:/Users/chand/Desktop/CapStone/chest_xray/train/NORMAL"
TARGET_NORMAL_DIR = "C:/Users/chand/Desktop/CapStone/Dataset/Normal"
if not os.path.exists(TARGET_NORMAL_DIR):
    os.mkdir(TARGET_NORMAL_DIR)
    print("Normal Folder Created.")
```

```python
image_names = os.listdir(NORMAL_FILE_PATH)
image_names
```

```python
random.shuffle(image_names)
image_names
```

```python
for i in range(cnt):
    image_name = image_names[i]
    image_path = os.path.join(NORMAL_FILE_PATH, image_name)
    target_path = os.path.join(TARGET_NORMAL_DIR, image_name)
    shutil.copy2(image_path,target_path)
    print("Copying image ",i)
```
